```python
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate
from matplotlib import cm
```

```python
import algo
```

```python
ww = float(input("entrez valeur de omega [0..100] "))
eps = float(input("entrez valeur de epsilon [0..15] "))

ZoomX = float(input("entrez valeur de l'intervale x [0..10]"))
ZoomY = float(input("entrez valeur de l'intervale |y| [0..10]"))

y0 = float(input("entrez valeur de y0 ["+ str(-ZoomY)+".."+str(ZoomY)+"] "))

a = float(input("entrez valeur de a [-40..-1] "))
b = float(input("entrez valeur de b [1..40] "))
p = int(input("diviser en p intervale [10..100] "))

eps = complex(0,eps)
```

```python
dx = 0.01
dy = 0.01
listX = np.arange(dx,ZoomX+dx*0.5,dx)
listY = np.arange(-ZoomY,ZoomY+dy*0.5,dy)
```

```python
resTrap  = np.zeros((len(listX),len(listY)),dtype=complex)
resGauss = np.zeros((len(listX),len(listY)),dtype=complex)
resHerm  = np.zeros((len(listX),len(listY)),dtype=complex)
```

```python
const = ww**2+eps
div_2pi = 0.5/np.pi

for ix,x in enumerate(listX):
    resy = []
    # Completer ICI  
    myf = lambda xi :  np.exp(1.0j*(listX[ix]*np.sqrt(const-xi**2) + xi*(listY-y0) ))
    
    mydf = lambda xi : myf(xi)*1.0j*(-listX[ix] * xi/np.sqrt(const-xi**2) + (listY-y0))

    resQuadTrap  = algo.quadTrap(a,b,p,myf)
    resQuadGauss = algo.quadGauss(a,b,p,myf)
    resQuadHerm  = algo.quadHerm(a,b,p,myf,mydf)
    
    resTrap[ix,:]  =  resQuadTrap
    resGauss[ix,:] =  resQuadGauss
    resHerm[ix,:]  =  resQuadHerm

    
    
resTrap  = div_2pi * np.array(resTrap).T
resGauss = div_2pi * np.array(resGauss).T
resHerm  = div_2pi * np.array(resHerm).T
```

```python
ptX,ptY = np.meshgrid(listX,listY)


fig, axs = plt.subplots(nrows=1,ncols=3,figsize=(15,4))

axs[0].pcolor(ptX,ptY,np.real(resTrap),cmap = cm.jet)
axs[0].axis('equal')

axs[1].pcolor(ptX,ptY,np.real(resGauss),cmap = cm.jet)
axs[1].axis('equal')

axs[2].pcolor(ptX,ptY,np.real(resHerm),cmap = cm.jet)
axs[2].axis('equal')

fig.suptitle("omega = "+str(ww)+" epsilon = "+str(np.abs(eps))+" a = "+str(a)+" b = "+str(b))
```
