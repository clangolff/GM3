```python
import matplotlib.pyplot as plt
import numpy as np
from tabulate import tabulate
from matplotlib import cm
```

```python
import algo
import data
```

## Test interpolation d'une fonction

```python
# Intervalle d'int√©gration
# Modifier ICI :
a=-1
b=1

# p-sous intervalles pour la formule de quadrature (composite) :
# Modifier ICI :
p = 20
h = (b-a)/p



f = data.f
df = data.df
```

```python
################# AFFICHAGE ######################

xx = np.arange(a,b,max(h*0.01,1e-3))
yy = f(xx)

fig, axs = plt.subplots(nrows=1,ncols=3,figsize=(15,4))

for i in range(3) :
    axs[i].plot(xx,yy, '-',color='#2ca02c',linewidth=2.0,label="f")
    
    # positionne la vue sur le bon intervalle
    axs[i].set_xlim(a-0.1,b+0.1)
    axs[i].set_ylim(min(min(yy),0.0)-0.1,max(max(yy),0.0)+0.1)

    # positionne l'axe horizontale du bas au milieu
    axs[i].spines['bottom'].set_position('zero') 
    
    # efface l'axe horizontale en haut et l'axe verticale √† droite
    axs[i].spines['right'].set_color('none')
    axs[i].spines['top'].set_color('none')


    
                                        ##### Trap√®ze ##### 
xhT = np.array([a +i*h for i in range(p)])
yhT = f(xhT)

axs[0].plot(xhT,yhT,'x',color='#ff7f0e')
axs[0].plot(xhT,yhT,'--',color='#1f77b4',linewidth=2.0,label="interp")
axs[0].plot([xhT[-1],b],[f(xhT[-1]),f(b)],'--',color='#1f77b4',linewidth=2.0)


x = a

c1 = (-1. + np.sqrt(3)) / np.sqrt(12)  
c2 = ( 1. + np.sqrt(3)) / np.sqrt(12)  

while(x < b-1e-8):

                                        ##### Gauss #####
            
        
        # xhG est une liste de 11 pnts entre xi et xi+1 
    xhG = np.arange(x,x+h+h*0.01,h*0.1)
        
        # points interpolants 
    x1G = x + c1 * h 
    x2G = x + c2 * h
        
        # √©valuations des points interpolants pour le calcul de la quadrature
    fx1G = f(x1G)
    fx2G = f(x2G)
        
        # √©valuation de la liste de point dans la base de Lagrange ! ( p(x) = f(x1)*L_x1 + f(x2)*L_x2 ) 
    yhG = fx1G * (xhG-x2G) / (x1G-x2G) + fx2G * (xhG-x1G)/ (x2G-x1G)

    axs[1].plot(xhG,yhG,'--',color='#1f77b4',linewidth=2.0)
    axs[1].plot([x1G,x2G],[fx1G,fx2G],'x',color='#ff7f0e')
      
                                        
                                        ##### Hermite #####
          
        # xhG est une liste de 11 pnts entre xi et xi+1 
    xhH = np.arange(x,x+h+h*0.01,h*0.1)
        
        # points interpolants
    x0H = x           # d√©but
    x1H = x+h*0.5     # milieu
    x2H = x+h         # fin
 
    df0 = data.df(x0H)
    df2 = data.df(x2H)

    yhH = f(x1H) + (xhH - x1H)*0.5*(df0 + df2) + ((xhH-x1H)**2)*0.5*(df2 - df0)/h

    axs[2].plot(xhH,yhH,'--',color='#1f77b4',linewidth=2.0)
    axs[2].plot([x1H],[f(x1H)],'x',color='#ff7f0e')
 
    x = x+h

        
axs[0].legend()
axs[1].legend(["f","interp"])
axs[2].legend(["f","interp"])
axs[0].set_title('Trap√®ze')
axs[1].set_title('Gauss')
axs[2].set_title('Hermite')

plt.show()
```

## Analyse de convergence

```python
resQuadTrap = 0.0
resQuadGauss = 0.0
resQuadHerm = 0.0

#Pour f(x)=cos(2kpi*x)
#kk = 5.0
#valEx = 1/(2*kk*np.pi) * np.sin(1.5 * kk * np.pi)
#valEx = 0.75**3 / 3
valEx = 1/5*(np.arctan(5*b)-np.arctan(5*a))

list_p = [4,8,16,32,64,128]

list_err = []
list_res = []
for p in list_p:
    resQuadTrap  = algo.quadTrap(a,b,p,f)
    resQuadGauss = algo.quadGauss(a,b,p,f)
    resQuadHerm  = algo.quadHerm(a,b,p,f,df)

    res = [resQuadTrap,resQuadGauss,resQuadHerm]
    list_res.append(res)
    list_err.append(np.abs(res-valEx*np.ones((3,))))

    

list_err = np.array(list_err)
list_res = np.array(list_res)
list_p = np.array([list_p])

poly = np.polyfit(np.log10(list_p[0]),np.log10(list_err[:,0]),1)
```

```python
titreX = np.array([["p","err Trap√®ze","err Gauss","err Hermite"]])
t_err = np.concatenate((list_p.T,list_err),axis=1)
print(tabulate(np.concatenate((titreX,t_err)),tablefmt="fancy_grid"))
```

```python
titreX = np.array([["p","res Trap√®ze","res Gauss","res Hermite"]])
t_res = np.concatenate((list_p.T,list_res),axis=1)
print(tabulate(np.concatenate((titreX,t_res)),tablefmt="fancy_grid"))
```

```python
Tpoly = np.zeros((3,2))
fig, axs = plt.subplots(nrows=1,ncols=3,figsize=(15,4))

for i in range(3) :
    x = np.log10(list_p[0])
    y = np.log10(list_err[:,i])
    Tpoly[i] = np.polyfit(x,y,1)
    a,b = Tpoly[i,0],Tpoly[i,1]
        
    axs[i].plot(x,y,"+g-",label="log10(err)vslog10(p)")

    axs[i].set_xlabel("y ="+str(round(a,4)) +' x'+ str(round(b,4)))
    axs[i].legend()

axs[0].set_title('Trap√®ze')
axs[1].set_title('Gauss')
axs[2].set_title('Hermite')
    
```

## Resolution Helmotz

```python
dx = 0.01
dy = 0.01
listX = np.arange(dx,5+dx*0.5,dx)
listY = np.arange(-2.5,2.5+dy*0.5,dy)

ww = 15.0
eps = 0.1j
y0 = 0.0;

p = 100
resTrap  = np.zeros((len(listX),len(listY)),dtype=complex)
resGauss = np.zeros((len(listX),len(listY)),dtype=complex)
resHerm  = np.zeros((len(listX),len(listY)),dtype=complex)

const = ww**2+eps
div_2pi = 0.5/np.pi


for ix,x in enumerate(listX):
    resy = []
    # Completer ICI  
    myf = lambda xi : np.exp(1.0j*(listX[ix]*np.sqrt(const-xi**2) + xi*(listY-y0) ))
    
    mydf = lambda xi : myf(xi)*1.0j*(-listX[ix] * xi/np.sqrt(const-xi**2) + (listY-y0))


    # Modifier ICI :
    a = -20.0
    b = 20.0

    resQuadTrap  = algo.quadTrap(a,b,p,myf)
    resQuadGauss = algo.quadGauss(a,b,p,myf)
    resQuadHerm  = algo.quadHerm(a,b,p,myf,mydf)
    
    resTrap[ix,:]  =  resQuadTrap
    resGauss[ix,:] =  resQuadGauss
    resHerm[ix,:]  =  resQuadHerm

    
    
resTrap  = div_2pi * np.array(resTrap).T
resGauss = div_2pi * np.array(resGauss).T
resHerm  = div_2pi * np.array(resHerm).T
```

```python
ptX,ptY = np.meshgrid(listX,listY)

fig, axs = plt.subplots(nrows=1,ncols=3,figsize=(15,4))

axs[0].pcolor(ptX,ptY,np.real(resTrap),cmap = cm.jet)
axs[0].axis('equal')

axs[1].pcolor(ptX,ptY,np.real(resGauss),cmap = cm.jet)
axs[1].axis('equal')

axs[2].pcolor(ptX,ptY,np.real(resHerm),cmap = cm.jet)
axs[2].axis('equal')

fig.suptitle("omega = "+str(ww)+" epsilon = "+str(np.abs(eps))+"a="+str(a)+"b="+str(b))
fig.savefig("omega = "+str(ww)+" epsilon = "+str(np.abs(eps))+"a="+str(a)+"b="+str(b)+".png")
```

```python
fig, ax = plt.subplots(nrows=1,ncols=3,figsize=(15,5),subplot_kw={"projection": "3d"})

surf = ax[0].plot_surface(ptX, ptY, np.real(resTrap), cmap=cm.jet, linewidth=0, antialiased=False, shade=False)
surf = ax[1].plot_surface(ptX, ptY, np.real(resGauss), cmap=cm.jet, linewidth=0, antialiased=False, shade=False)
surf = ax[2].plot_surface(ptX, ptY, np.real(resHerm), cmap=cm.jet, linewidth=0, antialiased=False, shade=False)
```

```python
dx = 0.01
dy = 0.01
listX = np.arange(dx,7+dx*0.5,dx)
listY = np.arange(-5,5+dy*0.5,dy)

ww = 10.0
eps = 1.0j

y0_0 = 1.
y0_1 = -1.

p = 50
resTrap1  = np.zeros((len(listX),len(listY)),dtype=complex)
resGauss1 = np.zeros((len(listX),len(listY)),dtype=complex)
resHerm1  = np.zeros((len(listX),len(listY)),dtype=complex)


resTrap2  = np.zeros((len(listX),len(listY)),dtype=complex)
resGauss2 = np.zeros((len(listX),len(listY)),dtype=complex)
resHerm2  = np.zeros((len(listX),len(listY)),dtype=complex)

const = ww**2+eps
div_pi = 1./np.pi
T = 0.1

for ix,x in enumerate(listX):
    resy = []
    # Completer ICI  
    myf1 = lambda xi : np.sinc(xi*T)*np.exp(1.0j*(listX[ix]*np.sqrt(const-xi**2) + xi*(listY-y0_0) ))
    
    mydf1 = lambda xi : myf1(xi)*(-listX[ix] * xi/(const-xi**2) + (listY-y0_0))

    myf2 = lambda xi : np.sinc(xi*T)*np.exp(1.0j*(listX[ix]*np.sqrt(const-xi**2) + xi*(listY-y0_1) ))
    
    mydf2 = lambda xi : myf2(xi)*(-listX[ix] * xi/(const-xi**2) + (listY-y0_1))

    # Modifier ICI :
    a = -30.0
    b = 30.0

    resQuadTrap1  = algo.quadTrap(a,b,p,myf1)
    resQuadGauss1 = algo.quadGauss(a,b,p,myf1)
    resQuadHerm1  = algo.quadHerm(a,b,p,myf1,mydf1)
    
    resQuadTrap2  = algo.quadTrap(a,b,p,myf2)
    resQuadGauss2 = algo.quadGauss(a,b,p,myf2)
    resQuadHerm2  = algo.quadHerm(a,b,p,myf2,mydf2)
    
    
    resTrap1[ix,:]  = resQuadTrap1
    resGauss1[ix,:] = resQuadGauss1
    resHerm1[ix,:] = resQuadHerm1
    
    
    resTrap2[ix,:]  = resQuadTrap2
    resGauss2[ix,:] = resQuadGauss2
    resHerm2[ix,:] = resQuadHerm2

    
resTrap  = div_pi * T * (np.array(resTrap1).T + np.array(resTrap2).T)
resGauss = div_pi * T * (np.array(resGauss1).T + np.array(resGauss2).T)
resHerm  = div_pi * T * (np.array(resHerm1).T + np.array(resHerm2).T)
```

```python
ptX,ptY = np.meshgrid(listX,listY)

fig, axs = plt.subplots(nrows=1,ncols=3,figsize=(15,4))

axs[0].pcolor(ptX,ptY,np.real(resTrap),cmap = cm.jet)
axs[0].axis('equal')

axs[1].pcolor(ptX,ptY,np.real(resGauss),cmap = cm.jet)
axs[1].axis('equal')

axs[2].pcolor(ptX,ptY,np.real(resHerm),cmap = cm.jet)
axs[2].axis('equal')
```
